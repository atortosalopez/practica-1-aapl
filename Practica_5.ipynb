 2.   Cree un DataFrame fruit_sales que coincida con el siguiente diagrama:

|   | Apples  | Bananas   |
| -------  | ------- | -------- |
| 2017 Sales  | 35   | 21    |
| 2018 Sales  | 41   | 34    |

 3.   Crea una variable ingredients con una Series que se vea así:

Flour  4 cups    
Milk   1 cup     
Eggs   2 large   
Spam   1 can     
Name: Dinner, dtype: object

```python
# Crear la Serie ingredients
ingredients = pd.Series(
    ["4 cups", "1 cup", "2 large", "1 can"],
    index=["Flour", "Milk", "Eggs", "Spam"],
    name="Dinner"
)

ingredients
```
